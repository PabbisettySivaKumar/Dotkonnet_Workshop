### 2. Summary Statistics

.describe() for numeric columns.

Count missing values per column.

Fill missing numeric values with the mean.

```python
df.describe().round(3)
```

```python
for col, missing in df.isnull().sum().items():
    print(f'No of Missing values in {col}: {missing}')
```

```python
df.fillna(df.mean(numeric_only=True), inplace= True)
```

### 3. Filter & Sort

Filter rows by a numeric condition (e.g., Age > 30).

Sort dataset by a column in descending order.

```python
filter_df= df[df['quality']>6]
filter_df
```

```python
df.sort_values(by= ['pH', 'quality'], ascending=[True, False])
```

### 4. Group & Aggregate

Group by a categorical column, calculate mean of a numeric column.

```python
alcohol_avg= df.groupby('quality')['alcohol'].mean()
alcohol_avg
```

### 5. Visualize

Create a histogram for a numeric column.

Create a bar chart of group averages.

```python
for col in df.columns:
    plt.figure(figsize=(4,4))
    plt.hist(df[col], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('frequency')
    plt.show()
```
